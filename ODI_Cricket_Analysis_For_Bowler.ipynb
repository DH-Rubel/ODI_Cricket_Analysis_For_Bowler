# **Markdown Cell for explaining the meaning of each column**
 


*   **Span:** Total period of time playing a player.
*   **Mat:** The Number of match played.
*   **Inns:** The number of innings in which the bowler actually bowled.
*   **Balls:** The number of balls bowled.
*   **Runs:** Total number of runs conceded.
*   **Wkts:** The number of wickets taken.
*   **Ave:** The number of runs a bowler has conceded per wicket taken.
*   **Econ:** The average number of runs conceded per over. (Econ = Runs/Overs bowled).

*   **SR:** Strike rate. The average number of balls bowled per wicket taken. (SR = Balls/W)
*   **4:** The number of innings in which the bowler took at least four wickets.
*  **5:** The number of innings in which the bowler took at least five wickets.

# **Renaming the column names**

```python
df = df.rename(columns={'Mat':'Match', 
                        'Inns':'Innings',
                        #'Balls': 'The number of balls bowled',
                        #'Runs': 'The number of runs conceded',
                        'Wkts': 'Wickets',
                        #'Wkts':'The_number_of_wikets_taken',
                        'Ave': 'Bowling_average',
                        'Econ':'Economy_rate',
                        'SR': 'Strike_Rate'})

display(df.head())
```

# **Spliting the player column into two different columns named, player_Name & Country.**

```python
df[["Player_Name", "Country"]] = df['Player'].str.split("(", expand=True)

display(df.head())
```

# **Omitting the parentheses of the last part of first bracket. i.e.  )**

```python
df['Country'] = df['Country'].str.replace(")", "")

display(df.head())
```

```python
df['Country'].value_counts()
```

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_ICC'] = df['Country'].apply(icc_check)
display(df.head())
```

```python
#display(df.head(37))
#display(df.tail(40))
display(df.sample(8))
```

# **Finding the number of players played for ICC**

```python
print(df['played_for_ICC'].value_counts())
```

**Here, Thirteen players played for ICC**

```python
df['Country'] = df['Country'].str.replace("ICC/", "")
df['Country'] = df['Country'].str.replace("/ICC", "")
df['Country'] = df['Country'].str.replace("Asia/", "")
df['Country'] = df['Country'].str.replace("Afr/", "")
display(df.head(10))
#display(df.head(37))
#display(df.tail(40))
```

# **Spliting the span column into two different columns named, start_year end_year and changing the data type string to integer.**

```python
df[['start_year', 'end_year']] = df['Span'].str.split("-", expand=True)
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')
print(df.info())
```

# **Finding the player who played for the longest period of time**

```python
df['years_played'] = df['end_year'] - df['start_year']+1
#display(df.head(37))
#display(df.tail(40))
df.sort_values(by='years_played', ascending = False).head(8)
#display(df.tail(10))
```
